# Customer review sentiment analysis for British airways

### Exploratory data analysis

```python
# Importing libraries
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
df = pd.read_csv('data.csv', encoding='ISO-8859-1', on_bad_lines='skip')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
print(len(df))
df.isnull().sum()
```

```python
# Bad data
df[df.review.isnull() == False]
```

```python
# Deleting bad data (rows with mismatched data in columns)
clean_df = df.drop(index=df[df.review.isnull() == False].index)

# Now we will drop the review column as it only has null values
clean_df.drop(columns=['review'], inplace=True)
clean_df.head()
```

## Changing the date to pandas datetime format

```python
clean_df.date = pd.to_datetime(clean_df.date, format="mixed")
clean_df.head(2)
```

## Plotting time series plots for rating and recommended 

```python
data = clean_df[['rating', 'date', 'Recommended']]
print(data.isna().sum())

# Cleaning all the NaN values
data = data.dropna()
data.isna().sum()
```

```python
print(len(data[data.Recommended == "no"]))
len(data[data.Recommended == "yes"])
```

```python
data['year'] = data['date'].dt.year

# group data by year and Recommended values
grouped = data.groupby(['year', 'Recommended'])['Recommended'].count()

# convert the grouped data to dataframe
grouped_df = grouped.unstack()

# plot the data
grouped_df.plot(kind='bar', stacked=False)

# add labels and title
plt.title('Total Recommendations by Year')
plt.xlabel('Year')
plt.ylabel('Total Recommendations')

plt.show()
# print(grouped_df)
```

## Observations -
#### Majority of recommendations have been negative since year 2016

```python
sentiment_df = clean_df.drop(clean_df[clean_df.header == "British Airways customer review"].index)
sentiment_df = sentiment_df[['header', 'date']]
sentiment_df.isnull().sum()
sentiment_df.dropna(inplace=True)
sentiment_df.isnull().sum()
```

```python
sentiment_df['year'] = sentiment_df.date.dt.year
```

```python
def get_sentiments(sentences):
    sentiments = []
    for sentence in sentences:
        blob = TextBlob(sentence)
        sentiment = blob.sentiment.polarity
        
        #Custom formula to convert polarity 
        # 0 = (Negative) 1 = (Neutral) 2=(Positive)
        sentiment = round(( sentiment + 1 ) * 3 ) % 3
        
        if sentiment == 2:
            return 'Positive'
        elif sentiment == 0:
            return 'Negative'
        else:
            return 'Neutral'
```

```python
sentiment_df['sentiment'] = sentiment_df['header'].apply(get_sentiments)
sentiment_df.head()
```
